1. Check whether given year is a leap year or not
2. Calculate income tax for Indian person age<60 and assume no tax benifits and all income is from salary FY - 24-25

```python
# Take input from user
year = int(input("Please enter the year : "))

# Apply if else conditon
if (year%4 == 0 and year%100 != 0) or (year%400 == 0):
    print(f'The given year {year} is a LEAP YEAR')
else:
    print(f'The given year {year} is NOT a leap year')
```

## 2. Calculate income tax for Indian person age<60 and assume no tax benifits and all income is from salary FY - 24-25

1. On annual income a standard deduction of 50000 is done.
2. Upto 7 lakhs Taxable income (7.5 lakh annual income) No tax is applied
3. After calculating tax 4% cess on tax amount is always applied
4. Final Tax = Tax + 4% cess on tax amount

```python
# Take annual income from user as input
income = float(input("Please enter your annual income in INR : "))

# Apply standard deduction of 50000
ti = income - 50000

# Create if else statements for tax slabs
if ti<300000:
    tax = 0
elif ti>=300000 and ti<600000:
    tax = (5/100)*(ti - 300000)
elif ti>=600000 and ti<900000:
    tax = 15000 + (10/100)*(ti-600000)
elif ti>=900000 and ti<1200000:
    tax = 15000 + 30000 + (15/100)*(ti - 900000)
elif ti>=1200000 and ti<1500000:
    tax = 15000 + 30000 + 45000 + (20/100)*(ti - 1200000)
else:
    tax = 15000 + 30000 + 45000 + 60000 + (30/100)*(ti - 1500000)

# Tax slabs are applicable only to taxable income > 700000
if ti <= 700000:
    tax = 0

# Calculate cess on tax of 4%
cess = (4/100)*tax

# Total Tax
total_tax = tax + cess

# Print the results
print(f'Annual Income : {income:.2f} INR')
print(f'Taxable Income : {ti:.2f} INR')
print(f'Tax amount : {tax:.2f} INR')
print(f'Cess 4% of Tax : {cess:.2f} INR')
print(f'Total Tax : {total_tax:.2f} INR')
```

### Find out squares of first n natural numbers, take n as input from user. I want output as a list

```python
list(range(1, 10))
```

```python
list(range(1, 11))
```

```python
# Take input from user
n = int(input("Please enter a positive number : "))

# Intiaize blank list for storing squares
s = []

# Apply for loop
for i in range(1, n+1):
    s.append(i**2)

# Print the list of squares
print(f'The squares of first {n} natural numbers is : {s}')
```
